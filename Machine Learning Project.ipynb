```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import mltools as ml

# seed for reproducing same results
seed = 9
np.random.seed(seed)

X = np.genfromtxt('data/X_train.txt', delimiter=None)
Y = np.genfromtxt('data/Y_train.txt', delimiter=None)
X_validation = np.genfromtxt('data/X_test.txt', delimiter=None)

X,Y = ml.shuffleData(X, Y)

# X2 = ml.transforms.fpoly(X, 2, bias = True)
# X_validation2 = ml.transforms.fpoly(X_validation, 2, bias = True)

# Splitting the dataset into the Training set and Test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)



# X_train2, params = ml.rescale(X_train2)
# X_test2, _ = ml.rescale(X_test2, params)

# Y_dummy = np.zeros((len(Y_train), 2))

# for i, c in enumerate(Y_train):
#     if c==0:
#         Y_dummy[i,0] = 1
#     else:
#         Y_dummy[i, 1] = 1



# Feature Scaling

# sc = StandardScaler()
# X_train = sc.fit_transform(X_train)
# X_test = sc.transform(X_test)
# X_validation = sc.transform(X_validation)
# X = sc.transform(X)
```

```python
# create the model

# model = Sequential()
# model.add(Dense(150, input_shape=(120,), activation='relu'))
# model.add(Dense(70, activation='relu'))
# model.add(Dense(2, activation='sigmoid'))

classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 20, init = 'uniform', activation = 'relu', input_dim = 14))
# Adding the second hidden layer
classifier.add(Dense(output_dim = 18, init = 'uniform', activation = 'relu'))
# Adding the third hidden layer
classifier.add(Dense(output_dim = 18, init = 'uniform', activation = 'relu'))
# Adding the fourth hidden layer
classifier.add(Dense(output_dim = 16, init = 'uniform', activation = 'relu'))
# Adding the fifth hidden layer
classifier.add(Dense(output_dim = 16, init = 'uniform', activation = 'relu'))
# Adding the sixth hidden layer
classifier.add(Dense(output_dim = 10, init = 'uniform', activation = 'relu'))
# Adding the seventh hidden layer
classifier.add(Dense(output_dim = 10, init = 'uniform', activation = 'relu'))
# Adding the eight hidden layer
classifier.add(Dense(output_dim = 8, init = 'uniform', activation = 'relu'))
# Adding the ninth hidden layer
classifier.add(Dense(output_dim = 8, init = 'uniform', activation = 'relu'))
# Adding the tenth hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
# Adding the eleventh hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

# compile the model
# model.compile(loss='categorical_crossentropy', optimizer='SGD', metrics=['accuracy'])
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# fit the model
classifier.fit(X, Y, batch_size=50, nb_epoch=200, verbose=1)
```

```python
# evaluate the model
scores = classifier.evaluate(X_test, Y_test)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
